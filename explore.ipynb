```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import altair as alt
```

```python
# identify user attributes
users = pd.read_csv("data/user_targets.csv")
users.columns
```

```python
# identify playlist attributes
playlists = pd.read_csv("data/df_playlists.csv")
for c in playlists.columns:
    print(c)
```

```python
# extract a list of genres column names
playlists_cols = list(playlists.columns)
genres_start = playlists_cols.index("pop")
genres_cols = playlists_cols[genres_start:]

# identify whether a genre exists in a playlist (binary genres df)
genres_df = playlists[genres_cols]
bin_genres = (genres_df[genres_cols] > 0).astype(int)

# identify the top 10 most popular genres
bin_genres.mean().nlargest(5)
```
